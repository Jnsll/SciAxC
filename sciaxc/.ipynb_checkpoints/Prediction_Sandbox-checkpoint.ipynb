```python
def get_random_test_site_and_corresponding_known_sites(df_optimum_rate_global):
    df_test_site = df_optimum_rate_global.sample()
    test_site_number = int(df_test_site["SiteNumber"])
    print("Site number:", int(df_test_site["SiteNumber"]))
    index_site_test = df_test_site.index[0]
    df_known_points = df_optimum_rate_global
    df_known_points = df_known_points.drop(index_site_test)
    df_known_points = df_known_points.reset_index(drop=True)
    return df_test_site, df_known_points, test_site_number
```

```python
import numpy as np

def compute_distances_for_test_site(test_site_number, df_known_points, df_test_site):
    dict_distances_l2 = {}
    dict_distances_l2[test_site_number] = {}
    for site_index in df_known_points.index:
        dist = (np.linalg.norm(df_test_site.iloc[0,2:]-df_known_points.iloc[site_index,2:]))
        know_site_number = df_known_points.iloc[site_index,0]
        dict_distances_l2[test_site_number][know_site_number] = dist
    return dict_distances_l2
```

```python
dict_ordered = {k: v for k, v in sorted(dict_distances_l2[test_site_number].items(), key=lambda item: item[1])}
```

```python
dict_ordered
```

```python
from more_itertools import take

def get_rates_nearest_neighbours(n, dict_ordered, df_known_points, df_test_site):
    n_items = take(n, dict_ordered.items())
    nearest_neighbours_sites = []
    rates_nearest_neighbours = []
    for site in range(len(n_items)):
        nearest_neighbours_sites.append(n_items[site][0])
    for site in nearest_neighbours_sites:
        #print(site)
        row = df_known_points.loc[df_known_points['SiteNumber'] == site]
        rate_nearest_neighbour = int(row["Rate"])
        rates_nearest_neighbours.append(rate_nearest_neighbour)
        print("Rate of nearest neighbour:", rate_nearest_neighbour)
    print("Test site Rate:", int(df_test_site["Rate"]))
    return rates_nearest_neighbours, int(df_test_site["Rate"])
```

```python
import statistics
def pipeline_for_n_neighbours(n):
    df_test_site, df_known_points, test_site_number = get_random_test_site_and_corresponding_known_sites(df_optimum_rate_global)
    dict_distances_l2 = compute_distances_for_test_site(test_site_number, df_known_points, df_test_site)
    dict_ordered = {k: v for k, v in sorted(dict_distances_l2[test_site_number].items(), key=lambda item: item[1])}
    rates_nearest_neighbours, rate_test_site = get_rates_nearest_neighbours(n, dict_ordered, df_known_points, df_test_site)
    #print(rates_nearest_neighbours)
    prediction = statistics.median(rates_nearest_neighbours)
    return prediction, rate_test_site
```

```python
number_repetitions = 5
n = 5
diffs = []
for rep in range(number_repetitions):
    prediction, rate_test_site = pipeline_for_n_neighbours(n)
    print("Prediction", prediction)
    diff_pred_real = rate_test_site - prediction
    print("Difference", diff_pred_real)
    diffs.append(diff_pred_real)

print(diffs)
    
```
